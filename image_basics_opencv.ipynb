#### Color channel difference between Matplotlib and OpenCV

There is a difference in how the color channels are arranged in OpenCV and Matplotlib.<br />
**Matplotlib** actually arranges the channels in the standard format, that is **RGB** format, whereas the **OpenCV** arranges them in the **BGR** format

However, we can change the arrangement in OpenCV with the method cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

#### Colors can also be changed to Grayscale or any other arrangement right when the image is being read by OpenCV

```python
img_gray = cv2.imread('kitten.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
img_gray.shape
```

```python
img_gray.max()
```

#### Even if we read the image in GRAYSCALE, if we plot it using matplotlib, it will show us some weird colors because it still thinks in the RGB format. We can avoid it using the 'cmap' argument.

```python
plt.imshow(img_gray, cmap = 'gray')
```

```python
plt.imshow(img_gray, cmap = 'magma')
```

### Drawing on Images

Let us create a blank black image to try drawing shapes.

```python
blank_img = np.zeros(shape = (512, 512, 3), dtype = np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

#### Drawing a Rectangle

```python
cv2.rectangle(blank_img, pt1 = (384, 0), pt2 = (510, 150), color = (0, 255, 0), thickness = 10)
```

```python
plt.imshow(blank_img)
```

#### Drawing a square

```python
cv2.rectangle(blank_img, pt1 = (200, 200), pt2 = (300, 300), color = (0, 0, 255), thickness = 10)
```

```python
plt.imshow(blank_img)
```

#### Drawing a circle

```python
cv2.circle(blank_img, center = (100, 100), radius = 50, color = (255, 0 ,0), thickness = 8)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, center = (400, 400), radius = 50, color = (255, 0 ,0), thickness = -1)
plt.imshow(blank_img)
```

#### Drawing a line

```python
cv2.line(blank_img, pt1 = (0,0), pt2 = (512,512), color = (102, 255, 255), thickness = 5)
plt.imshow(blank_img)
```

#### Writing a Text

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text = 'Hello', org = (10, 500), fontFace = font, fontScale = 4, color = (255, 255, 255), thickness = 3, lineType = cv2.LINE_AA)
plt.imshow(blank_img)
```

#### Drawing a Polygon

```python
vertices = np.array([[100, 300], [170, 340], [300, 200], [400, 400]], dtype = np.int32)
```

```python
vertices
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1, 1, 2))
```

```python
pts
```

```python
pts.shape
```

```python
cv2.polylines(blank_img, [pts], isClosed = True, color = (255, 0, 0), thickness = 5)
plt.imshow(blank_img)
```
